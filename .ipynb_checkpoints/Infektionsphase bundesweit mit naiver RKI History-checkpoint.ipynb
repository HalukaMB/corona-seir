## Wöchentlicher Trend mit naiver History

```python
def group_by_federal_state(daily_diff):
    ags_names = {
        "01": "Schleswig-Holstein",
        "02": "Hamburg",
        "03": "Niedersachsen",
        "04": "Bremen",
        "05": "Nordrhein-Westfalen",
        "06": "Hessen",
        "07": "Rheinland-Pfalz",
        "08": "Baden-Württemberg",
        "09": "Bayern",
        "10": "Saarland",
        "11": "Berlin",
        "12": "Brandenburg",
        "13": "Mecklenburg-Vorpommern",
        "14": "Sachsen",
        "15": "Sachsen-Anhalt",
        "16": "Thüringen",
        "su": "Bundesweit" # column was named "sum_total"
    }
    
    by_federal_state = daily_diff.transpose()
    by_federal_state = by_federal_state.groupby(by_federal_state.index.str.slice(0,2)).sum()
    by_federal_state.index = [ags_names[ags] for ags in by_federal_state.index]
    
    return by_federal_state.transpose() 
```

```python
by_federal_state = group_by_federal_state(daily_diff_naive)
by_federal_state.head()
```

```python
by_federal_state.plot(figsize=(16,9))
```

```python
def weekly_trend(daily_diff, i=0):
    weekly_trend = pd.DataFrame(columns=['Aktuell', 'Vorwoche', 'Trend', 'Datentyp'], index=daily_diff.columns)
    
    for federal_state in daily_diff.columns:
        # last seven days
        current_week = daily_diff[federal_state].iloc[i-7:].sum()
        # one week ago until two weeks ago
        past_week = daily_diff[federal_state].iloc[i-14:i-7].sum()

        ratio = current_week / past_week
        diff = current_week - past_week

        #if we have a 110% or higher ratio, we label it as surging
        if ratio > 1.1:
            indicator="Steigend"
        #if we have a 90% or lower ratio, we label it as declining
        elif ratio < 0.9:
            indicator="Sinkend"
        #if we have a ratio between the two numbers, we say it is unchanged
        else:
            indicator="Gleichbleibend"

        #if we have a |diff| that is less than 5, we also label it as unchanged
        if -5 <= diff <= 5:
            indicator="Gleichbleibend"

        weekly_trend.loc[federal_state] = {
            'Aktuell': current_week,
            'Vorwoche': past_week,
            'Trend': indicator,
            'Datentyp': 'Wochentrend'
        }

    return weekly_trend
```

```python
weekly_trend_naive = weekly_trend(by_federal_state)
weekly_trend_naive
```

```python
daily_trend = []

for federal_state in by_federal_state.columns:
    for i, v in enumerate(by_federal_state[federal_state]):
        # we skip the first two weeks because that's how far we need to look into the past
        if i < 14:
            continue
            
        # seven days ago to `i`
        current_week = by_federal_state[federal_state].iloc[i-7:i].sum()
        # one week ago until two weeks ago
        past_week = by_federal_state[federal_state].iloc[i-14:i-7].sum()

        ratio = current_week / past_week
        diff = current_week - past_week

        #if we have a 110% or higher ratio, we label it as surging
        if ratio > 1.1:
            trend = "Steigend"
        #if we have a 90% or lower ratio, we label it as declining
        elif ratio < 0.9:
            trend = "Sinkend"
        #if we have a ratio between the two numbers, we say it is unchanged
        else:
            trend = "Unklar"

        #if we have a |diff| that is less than 5, we also label it as unchanged
        if -5 <= diff <= 5:
            trend = "Unklar"
            
        # we calculate the tipping point as follows:
        # - get trends between two week and one week ago
        # - get the trends from one week ago until today
        # - if both weeks have a clear trend (one kind of trend on at least 4 days that week)
        # - and if they differ
        # → 📈 tipping point
        # if len(daily_trend) >= 14:
        #    previous_week_trends = Counter([day['trend'] for day in daily_trend[-14:-7]])
        #    current_week_trends = Counter([day['trend'] for day in daily_trend[-7:]])
        #    
        #    previous_trend, previous_trend_n = previous_week_trends.most_common()[0]
        #    current_trend, current_trend_n = current_week_trends.most_common()[0]
        #    
        #    is_tipping_point = previous_trend_n >= 3 and current_trend_n >= 3 and previous_trend != current_trend
        # else:
        #    is_tipping_point = False
        
        daily_trend.append({
            'bundesland': federal_state,
            'datum': by_federal_state.index[i],
            'heute': v,
            'vorwoche': past_week,
            'aktuell': current_week,
            'trend': trend,
            # currently_unused:
            # 'wendepunkt': is_tipping_point
        })
        
daily_trend = pd.DataFrame(daily_trend).set_index(['bundesland', 'datum'])

# only keep the first of all tipping points
# daily_trend['wendepunkt'] = daily_trend['wendepunkt']
daily_trend
```

```python
# how many days to watch for a trend?
lookahead = 7
# how many days in `lookahead` days do we have to see one trend to judge it as prevalent?
minimum_trend_in_days = 5

trend_watch = {}

#for each federal state
for Bundesland in by_federal_state.columns:
    # print(Bundesland)
    
    #we create a list of the current trends and we start with "not enough data" as default
    #in case we do not have a trend that reaches the required threshold of minimum_trend_in_days
    current_trend ="Nicht Genügend Daten"
    current_trend_list=[]

    # print(len(daily_trend.loc[Bundesland]))
    #we iterate throudh the daily trends of this federal state
    for i, (_, row) in enumerate(daily_trend.loc[Bundesland].iterrows()):
        # if the trend is prevalent for more than `minimum_trend_in_days`, this is our current trend otherwise
        # it stays with the previous one. we note a trend at its start, watched over seven days
        if i < len(daily_trend.loc[Bundesland]) - lookahead:
            last7days_trends = daily_trend.loc[Bundesland][i:i+lookahead]["trend"]
            last7days_trends = Counter(trend for trend in last7days_trends)
            if last7days_trends['Steigend'] >= minimum_trend_in_days:
                current_trend = "Steigend"
            elif last7days_trends['Sinkend'] >= minimum_trend_in_days:
                current_trend = "Sinkend"
            
            #then we append this to a list of trends
            current_trend_list.append(current_trend)            
    
    # because of the "we need at least 7 days of data for a trend" restriction we set above, we note lacking data
    # towards the end and append it to our trend list
    current_trend_list = current_trend_list + (['Nicht Genügend Daten'] * lookahead)
    # then we take the weekly new cases for each federal state
    trend_watch[Bundesland] = daily_trend.loc[Bundesland]["aktuell"]/7
    # and our defined trend into a dictionary
    trend_watch[Bundesland + "_trend"] = current_trend_list

# finally we build the dataframe from the dictionaries
trend_watch = pd.DataFrame(trend_watch)

# and convert the DateTimeIndex to a column holding the date like YYYY-MM-DD; this is needed for datawrapper
trend_watch['Datum'] = trend_watch.index.strftime('%Y-%m-%d')
trend_watch = trend_watch.reset_index().drop('datum', axis='columns')
trend_watch.tail(n=21)
```
